```python
## Chart Input - 2
year = 2024
month = 2
day = 5
hour = 15
minute = 51
secs = 0
horary_number = 140
```

```python
## General Specs for location, ayanamsa and house_system
latitude, longitude, utc = 11.020085773931049, 76.98319647719487, "+5:30" ## Coimbatore
ayan = "Krishnamurti"
house_system = "Placidus" ## Default House for KP system
```

```python
horary_asc = horary_chart.get_horary_ascendant_degree(horary_number)
desired_asc = horary_asc["ZodiacDegreeLocation"]
sublord =  horary_asc["SubLord"]
```

```python
%%time
matched_time, vhd_hora_houses_chart, houses_data = horary_chart.find_exact_ascendant_time(year, month, day, utc, latitude, longitude, horary_number, ayan)
secs_final = round(matched_time.second + (matched_time.microsecond) / 1_000_000, 3)
```

```python
asc = houses_data[0]
print(f"Horary Number: {horary_number} || Horary Asc Details: {horary_asc}")
if matched_time:
    print(f'The Ascendant is at {asc.LonDecDeg}° on {matched_time}')

print(f"Rounded Decimal Seconds: {secs_final}")
pretty_data_table(houses_data)    
```

```python
vhd_hora = VedicHoroscopeData(year, month, day, hour, minute, secs, utc, latitude, longitude, ayan, house_system)
vhd_hora_planets_chart = vhd_hora.generate_chart()
planets_data = vhd_hora.get_planets_data_from_chart(vhd_hora_planets_chart, vhd_hora_houses_chart)
planets_data_filtered = [planet for planet in planets_data if planet.Object not in ["Chiron", "Syzygy", "Fortuna"]]
pretty_data_table(planets_data_filtered)
```

```python
planets_significators_table = vhd_hora.get_planet_wise_significators(planets_data, houses_data)
pretty_data_table(planets_significators_table)
```
